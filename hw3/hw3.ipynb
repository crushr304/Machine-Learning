# Case 1 : Using Ridge(L2) and Lasso(L1) to predict the price directly

### Model Training

```python
from sklearn.model_selection import train_test_split
data = all_data
```

```python
data['MSSubClass'] = data['MSSubClass'].astype(str)

data['OverallCond'] = data['OverallCond'].astype(str)

data['YrSold'] = data['YrSold'].astype(str)
data['MoSold'] = data['MoSold'].astype(str)
```

用平均值填补缺失值

```python
x = data.loc[:,data.columns!='SalePrice']
mean_cols=x.mean()
x = x.fillna(mean_cols)  #填充缺失值
```

将合并后的数据集的非连续值通过pd.get_dummies将数据转换成one-hot编码，防止test和train特征值描述不同而产生的矩阵维度不对应

```python
x_dum = pd.get_dummies(x)
```

```python
train_x = x_dum[:1460]
test_x = x_dum[1460:]
train_y = train_price
```

```python
train_y
```

划分训练集和验证集

```python
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size = 0.3,random_state = 1)
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
models=[Ridge(),Lasso()]
names=['Ridge','Lasso']
score_=[]
```

```python
def try_different_method(name,model):
    model.fit(x_train,y_train)
    model.predict(x_test)
    score = model.score(x_test,y_test)
    score_.append(str(score))
    print("Start " + name +'. \n')
```

```python
for name,model in zip(names,models):
    try_different_method(name,model)
```

```python
scores = pd.DataFrame({'methods':names,'score':score_})
scores.set_index('methods',inplace=True)
scores
```

### Model Testing

```python
result = {}

def Predict(name, model):
    ret = model.predict(test_x)
    result[name] = ret
    frame = pd.DataFrame({'SalePrice':ret})
    print(frame)
    plt.figure(figsize=(25,10))
    plt.plot(np.arange(1461,1461+len(validation_price)), validation_price['SalePrice'], 'yo-', label='true value')
    plt.plot(np.arange(1461,1461+len(validation_price)), ret, 'o-', label='predict value')
    plt.xlabel("Houses' ID")
    plt.ylabel('SalePrice Prediction')
    plt.title('The Pridiction of SalePrice.')
    plt.legend()
    print(name)
    plt.show()
```

```python
for name, model in zip(names,models):
    Predict(name,model)
```

```python
result
```

Save as the csv file

```python
for name, ret in result.items():
    CSV = pd.DataFrame({'SalePrice':ret})
    path = './hw2/%s.csv' % name
    CSV.to_csv(path)
```

# Case 2 : Discrete the price then use Logisticregression to predict

### Model Training

##### discrete the price
* using scatter to display the distripution of SalePrice

```python
plt.figure(figsize=(20,10))
plt.scatter(np.array(range(0,1460)),train_price)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(np.array(range(1460,2919)),validation_price.SalePrice)
```

* discrete the price with four layer
* {'cheap':below 160000 ; 'moderate':160000-180000 ; 'high':180000-200000 ; 'expensive':200000+}

```python
train_y2 = []
validation_y = []

def trans(Price,ls):
    for price in Price:
        if price<=170000:
            ls.append('cheap')
        elif price>170000 and price<=180000:
            ls.append('moderate')
        elif price>180000 and price<=190000:
            ls.append('high')
        else:
            ls.append('expensive')
```

```python
trans(train_price,train_y2)
trans(validation_price.SalePrice,validation_y)
```

* Labels of train after discreting

```python
train_y2[:10]
```

* Labels of validation after discreting

```python
validation_y[:10]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_x, train_y2, test_size = 0.3, random_state=1)
```

##### 使用 multiclass 的 OvO 多分类策略,分类器使用 LogisticRegression

```python
model = OneVsOneClassifier(LogisticRegression(penalty='l2', solver='liblinear', max_iter=1000, class_weight='balanced'))
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

##### 使用 multiclass 的 OvR 多分类策略,分类器使用 LogisticRegression

```python
model2 = OneVsRestClassifier(LogisticRegression(penalty='l2', solver='liblinear', max_iter=1000, class_weight='balanced'))
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

##### 采用L2正则化和multinomial模式，multinomial即直接做多分类

```python
model3 = LogisticRegression(penalty='l2', solver='lbfgs', multi_class='multinomial', max_iter=1000, class_weight='balanced')
model3.fit(x_train, y_train)
y_pred = model3.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

### Model Testing

```python
price_pred = model3.predict(test_x)
accuracy_score(validation_y, price_pred)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(np.array(range(1460,2919)), validation_y)
plt.scatter(np.array(range(1460,2919)), price_pred, c='y')
plt.show()
```
